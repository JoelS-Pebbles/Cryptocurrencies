```python

# Initial imports
import pandas as pd
import numpy as np
import hvplot.pandas
from path import Path
import plotly.express as px
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
from path import Path
from sklearn.preprocessing import StandardScaler

```

```python
# Load the crypto_data.csv dataset.
# Load data
file_path =r"C:\Users\12109\Documents\Data boot camp\Class Work\Module 18\crypto_data.csv"
crypto_df = pd.read_csv(file_path, index_col=0)
print(crypto_df.shape)
crypto_df.head(10)
```

```python
# Keep all the cryptocurrencies that are being traded.
crypto2_df = crypto_df[(crypto_df["IsTrading"] == True)]
print(crypto2_df.shape)
crypto2_df.head(10)
```

```python
# Keep all the cryptocurrencies that have a working algorithm.
crypto3_df = crypto2_df.dropna(subset=["Algorithm"])
print(crypto3_df.shape)
crypto3_df.head(10)
```

```python
crypto4_df = crypto3_df.drop(columns=["IsTrading"])
print(crypto4_df.shape)
crypto4_df.head()
```

```python
crypto5_df = crypto4_df.dropna()
print(crypto5_df.shape)
crypto5_df.head(10)
```

```python
# Keep the rows where coins are mined.
# crypto6_df = crypto5_df.loc[crypto5_df.TotalCoinsMined !=0]
crypto6_df = crypto5_df.drop(crypto5_df[crypto5_df.TotalCoinsMined <= 0].index)
print(crypto6_df.shape)
crypto6_df.head(8)
```

```python
# Create a new DataFrame that holds only the cryptocurrencies names.
CoinName_df = crypto6_df.drop(columns= ["Algorithm", "ProofType", "TotalCoinsMined", "TotalCoinSupply"])
print(CoinName_df.shape)
CoinName_df.head()
```

```python
# Drop the 'CoinName' column since it's not going to be used on the clustering algorithm.
crypto_df = crypto6_df.drop(columns =["CoinName"])
print(crypto_df.shape)
crypto_df.head(10)
```

```python
# Use get_dummies() to create variables for text features.
crypto8_df = pd.get_dummies(crypto_df, columns = ["Algorithm", "ProofType"])
print(crypto8_df.shape)
crypto8_df.head(5)
```

```python
# Standardize the data with StandardScaler().
crypto_scaled = StandardScaler().fit_transform(crypto8_df)
crypto_scaled[0:5]
```

```python
# Using PCA to reduce dimension to three principal components.
pca = PCA(n_components=3)
crypto_pca = pca.fit_transform(crypto_scaled)
crypto_pca
```

```python
crypto_pca_df = pd.DataFrame(data=crypto_pca, columns=["PC_1", "PC_2", "PC_3"], index = crypto_df.index)
print (crypto_pca_df.shape)
crypto_pca_df.head(10)
```

```python
inertia=[]
k = list(range(1, 11))
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(crypto_pca_df)
    inertia.append(km.inertia_)
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", xticks=k, title="Elbow Curve")
```

```python
model = KMeans(n_clusters=4, random_state=0)

# Fit the model
model.fit(crypto_pca_df)

# Predict clusters
predictions = model.predict(crypto_pca_df)
print(predictions.shape)
predictions
```

```python
clustered_df = pd.concat([crypto_df, crypto_pca_df], axis=1)

clustered_df = pd.concat([clustered_df, CoinName_df], axis=1)

clustered_df["class"] = model.labels_

print(clustered_df.shape)
clustered_df.head(8)
```

```python
fig = px.scatter_3d(
    clustered_df, 
    x="PC_1", 
    y="PC_2", 
    z="PC_3", 
    color="class", 
    hover_name="CoinName", 
    hover_data=["Algorithm"], 
    symbol="class", 
    width=800
)
fig.update_layout(legend=dict(x=0,y=1))
fig.show()
```
